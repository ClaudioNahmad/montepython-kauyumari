```python
how_many_chains=len(chain_root)
#for i in chain_root:
#    print i

R1=[]


for i in chain_root:
    
    #these lines read .h_info (for R-1 values) and .v_info (for names of parameters) 
    r1=np.genfromtxt(i+'.h_info', skip_header=1, delimiter=':',dtype=None)
    nam=np.genfromtxt(i+'.v_info', skip_header=1, delimiter=':',dtype=None)
    
    #these lines convert the string 'r1'...
    v=r1[0,1].tostring()                       #array with just R-1numbers.(numbers as strings)
    f=v.split('\t')                            #float list with R-1 numbers as floats
    f.pop()                                    #removes last element of list (because it appears "empty")
    
    #these line converts 'nam' into a list of names
    names=nam[:,0]
    
    #print f
    #print names
    
    R1.append(f)

#print(R1)


R1[0]
test=R1[0]
#test2=R1[1]
#test3=R1[2]
#test4=R1[3]
#test5=R1[4]
```
